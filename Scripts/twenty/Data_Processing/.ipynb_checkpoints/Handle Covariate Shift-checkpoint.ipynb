```python
import pandas as pd
import numpy as np
import os
import time
from collections import defaultdict
from sklearn import linear_model
import math
```

```python
store_directory = "../Data/Symptoms_Demo/Covariate_balanced/"
directory = "../Data/Symptoms_Demo/Balanced_Data/"
```

```python
def read_file(filename):
    target = defaultdict()
    data = pd.read_csv(filename)
    cols = list(data.columns)
    #convert all the values from float to int
    data = data[cols].astype(int)
    cols = [i for i in cols if i!= 'virus']
    #create a unique value for each sample by concatenating the values in all columns
    data['set'] = data[cols].astype(str).apply(''.join, axis=1)
    return data
```

```python
def get_unique(df):
    columns_to_consider = list(df.columns)
    not_consider = ['set','target','virus']
    columns_to_consider = [i for i in columns_to_consider if i not in not_consider]
    print("Number of rows before removing the duplicates : ",df.shape[0])
    new_df = df.drop_duplicates(keep = 'first',subset = columns_to_consider)
    print("Number of unique rows : ",new_df.shape[0])
    return new_df
```

##### Balance nyumc and goviral

```python
#target dataset
filename_fluwatch = 'nyumc.csv'
data_fluwatch = read_file(directory+filename_fluwatch)
```

```python
data_fluwatch.head()
```

```python
data_fluwatch.shape
```

```python
#source file is honkong. The target dataset distribution is made similar to that of hongkong
filename_hk = 'hongkong.csv'
data_hk = read_file(directory+filename_hk)
data_hk.head()
```

```python
#adding a target variable to the datsets for dataset classificaton
data_fluwatch['target'] = 0
data_hk['target'] = 1
```

```python
def concatenate_datasets(d1,d2):
    return pd.concat([d1,d2])
```

```python
d1 = concatenate_datasets(data_fluwatch,data_hk)
```

```python
d1.head()
```

```python
print("Fuwatch sample points : ",data_fluwatch.shape[0])
print("Hongkong sample points : ",data_hk.shape[0])
print(d1.shape[0])
```

```python
features = d1.drop(['set','target','virus'],axis = 1)
lm = linear_model.LogisticRegression()
lm.fit(features,d1['target'])
```

```python
#get unique columns in nyumc:
unique_fluwatch = get_unique(data_fluwatch)
unique_fluwatch.head()
```

```python
sum(unique_fluwatch['virus'])
```

```python
coeff = lm.coef_.tolist()[0]
intercept = lm.intercept_.tolist()[0]
unique_features = unique_fluwatch.drop(['set','target','virus'],axis = 1)
unique_features.head()
```

```python
#convert the dataframe to a matrix
features = unique_features.as_matrix()
# predict on the examples
theta_x = list(np.dot(features,np.array(coeff)) + intercept)
#get the exponential of the prediction
exp_theta = [math.exp(i) for i in theta_x]
```

```python
new_exp_theta = []
for i in range(len(exp_theta)):
    if exp_theta[i] >= 0.1:
        new_exp_theta.append(exp_theta[i])
    else:
        new_exp_theta.append(0)
   
    
```

```python
len(exp_theta)
```

```python
len(new_exp_theta)
```

```python
print(new_exp_theta)
```

```python
min_pred = [i for i in new_exp_theta if i != 0]
print(min_pred)
```

```python
minimum_pred = min(min_pred)
```

```python
threshold = 1/minimum_pred
```

```python
threshold
```

```python
counts = data_fluwatch['set'].value_counts().to_dict()
```

```python
occur = []
for i in range(unique_fluwatch.shape[0]):
    x = list(unique_fluwatch.iloc[i,:]) 
    temp = new_exp_theta[i] * threshold * counts[x[-2]]
    occur.append(temp)
    
```

```python
unique_fluwatch['occur'] = occur
```

```python
unique_fluwatch.head()
```

```python
def duplicate_rows(dataframe,columns_to_consider=['fever', 'sorethroat', 'cough', 'muscle', 'headache', 'fatigue',
       'vomit', 'nausea', 'diarrhea', 'chills', 'sneeze',
       'shortness of breath', 'phlegm', 'blockednose', 'earache', 'leg pain',
       'runnynose', 'virus','age 0-4', 'age 5-15', 'age 16-44', 'age 45-64', 'age 65+',
       'male', 'female']):
    new_df = pd.DataFrame()
    for i in range(dataframe.shape[0]):
        ocurrence = int(dataframe.iloc[i]['occur'])
        temp = dataframe.iloc[i][:]
        if ocurrence!= 0:
            new_df = new_df.append([temp] * ocurrence)
    new_df = new_df.sample(frac=1).reset_index(drop=True)
    new_df = new_df[columns_to_consider]
    print(len(new_df))
    return new_df
```

```python
new_fluwatch = duplicate_rows(unique_fluwatch)
```
