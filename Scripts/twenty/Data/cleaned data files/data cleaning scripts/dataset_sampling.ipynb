### Mostly outdated functionality for sampling datasets
* Only use as a reference for how to sample

```python
dataset = pd.DataFrame.from_csv('gv_pre_sample.csv')
```

```python
# Loeb data
```

```python
dataset = dataset.fillna(0)
dataset = dataset.reset_index()
```

```python
dataset.apply(pd.value_counts).fillna(0)
```

```python
novirus = dataset[dataset['VIRUSDETECTED'] == 0]
virus = dataset[dataset['VIRUSDETECTED'] == 1]
```

```python
samplednovirus = novirus.sample(n=616,random_state=666)
sampledvirus = virus.sample(n=616,random_state=666)
```

```python
dataset = samplednovirus.append(sampledvirus)
```

```python
dataset_features = dataset.ix[:,3:13]
target = dataset.ix[:,14:].astype(str).astype(int)
```

```python
dataset_features[dataset_features != 0] = 1
```

```python
final_dataset = pd.concat([dataset_features,target], axis=1)
```

```python
final_dataset.to_csv("loeb_post_sample.csv")
```

```python
final_dataset.apply(pd.value_counts).fillna(0)
```

```python
pd.DataFrame(final_dataset.groupby(["VIRUSDETECTED","SinusproblemsSD"]).size())
```
