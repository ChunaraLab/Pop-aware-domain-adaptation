### Ensure that only the requried features are present

```python
required_features = ['fever',
                     'sorethroat',
                     'cough',
                     'muscle',
                     'headache',
                     'fatigue',
                     'vomit',
                     'nausea',
                     'diarrhea',
                     'chills',
                     'sneeze',
                     'shortness of breath',
                     'phlegm',
                     'blockednose',
                     'earache',
                     'leg pain',
                     'runnynose',
                     'age 0-4',
                     'age 5-15',
                     'age 16-44',
                     'age 45-64',
                     'age 65+',
                     'male',
                     'female',
                     'virus']
```

```python
def feature_selection(dataset,required_features = required_features):
    dataset_features = list(dataset.columns)
    #check if we have missed out any symptoms
    not_present = [i for i in required_features if i not in dataset_features]
    print("The features we have missed are : ",not_present)
    excess = [i for i in dataset_features if i not in required_features]
    print("The excess features are : ",excess)
    dataset.drop(excess,axis = 1,inplace = True)
    dataset = dataset[required_features]
    return dataset
```

```python
data_nyumc = feature_selection(data_nyumc)
```

```python
data_gv = feature_selection(data_gv)
```

```python
data_fw = feature_selection(data_fw)
```

```python
data_hk = feature_selection(data_hk)
```

```python
data_ht = feature_selection(data_ht)
```
