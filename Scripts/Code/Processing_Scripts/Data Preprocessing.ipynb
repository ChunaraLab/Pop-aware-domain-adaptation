```python
import pandas as pd
import numpy as np
import time
from collections import defaultdict
import os
import matplotlib.pyplot as plt
```

```python
def get_files(filename):
    data_files_ = []
    print("The data files are :\n")
    for fn in os.listdir(filename):
        if fn[-4:] == '.csv':
            data_files_.append(fn)
    print("\n",data_files_,"\n")
    print("Enter the number to analyze the file :")
    no = int(input())
    return data_files_[no]
```

```python
def read_file(filename):
    print("\n---------------------------------\n")
    print("The file being analyzed is :\t",filename)
    data = pd.read_csv(filename)
    print("\n---------------------------------\n")
#     print("The features of this data are :\n",data.columns)
#     print("\n---------------------------------\n")
    return data
```

```python
def select_dataset(filename_):
    data_file = get_files(filename_)
    filename = filename_+data_file
    data = read_file(filename)
    if 'muscles' in data.columns:
        data = data.rename(columns={'muscles': 'muscle'})
    return data
```

```python
GENDER_DEMOGRAPHICS = ['male','female']
AGE_DEMOGRAPHICS = ['age 0-4','age 5-15','age 16-44','age 45-64','age 65+']
```

```python
def check_demographics_in_dataset(dataset):
    not_present_gender = []
    not_present_age = []
    columns = dataset.columns
    
    #check if the gender demographics are as expected!
    for i in GENDER_DEMOGRAPHICS:
        if i not in columns:
            not_present_gender.append(i)
            
    #check if the age demographics are as expected!        
    for i in AGE_DEMOGRAPHICS:
        if i not in columns:
            not_present_age.append(i)
    
    return not_present_gender,not_present_age
```

```python
def duplicate_rows(dataframe,columns_to_consider):
    new_df = pd.DataFrame()
    for i in range(dataframe.shape[0]):
        new_df = new_df.append([dataframe.iloc[i][:]]*dataframe.iloc[i]['sum'],ignore_index=True)
    new_df = new_df.sample(frac=1).reset_index(drop=True)
    new_df = new_df[columns_to_consider]
    return new_df
```
