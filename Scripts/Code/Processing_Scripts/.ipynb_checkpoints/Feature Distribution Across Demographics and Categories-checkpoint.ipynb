### Same target variable for all the datasets

```python
#setting the target variable to virus if it is set to flu initially
def set_target(dataset):
    columns_ = dataset.columns
    if 'virus' not in columns_:
        print("Virus is not present in the columns!")
        dataset = dataset.rename(columns = {'flu':'virus'})
        print("New columns :\n")
        print(dataset.columns)
    else:
        print("Virus is already the target variable!")
    return dataset
```

### Get the distribution of symptoms for individual datasets (only for when they contribute to flu)

```python
def get_distribution(dataset,symptoms,presence_of = []):
    number_of_samples = dataset.shape[0]
    symptom_distribution = defaultdict()
    dataset = dataset.loc[dataset['virus'] == 1]
    if len(presence_of) == 0:
        for i in symptoms:
            if 1 in list(dataset[i]):
                frequency = round(dataset[i].value_counts()[1]/number_of_samples,5)
            else:
                frequency = 0
            symptom_distribution[i] = frequency
    else:
        dataset = dataset.loc[dataset[presence_of[0]] == 1]
        for i in symptoms :
            if 1 in list(dataset[i]):
                frequency = round(dataset[i].value_counts()[1]/number_of_samples,5)
            else:
                frequency = 0
            symptom_distribution[i] = frequency
       
    return symptom_distribution
```

```python
gv_sd = get_distribution(data_gv,symptoms,[])
print("GoViral!")
print(gv_sd)
```

```python
ht_sd = get_distribution(data_ht,symptoms,[])
print("Hutterite!")
print(ht_sd)
```

### Visualize the distribution for individual datasets

```python
def plot_symptom_distribution(symptom_frequency,
                             labels,
                             title):
    plt.figure(figsize=(16,8))
    
    ind = np.array([x for x,_ in enumerate(labels)])
    width = 0.6
    opacity = 0.2
    
    plt.bar(ind,symptom_frequency,width,alpha = opacity,color = 'b')
   
    plt.xticks(ind,labels,rotation = 45)
    plt.ylim(0,1)
    plt.ylabel('Frequency')
    plt.xlabel('Features')
    plt.title(title)
    plt.show()
```

```python
plot_symptom_distribution(nyumc_sd.values(),symptoms,'Symptom frequency distribution for NYUMC')
```

### Get the general distribution

```python
def get_general_distribution(dataset,symptoms,presence_of = []):
    number_of_samples = dataset.shape[0]
    symptom_distribution = defaultdict()
    dataset = dataset.loc[dataset['virus'] == 1]
    if len(presence_of) == 0:
        for i in symptoms:
            if 1 in list(dataset[i]):
                frequency = dataset[i].value_counts()[1]
            else:
                frequency = 0
            symptom_distribution[i] = frequency
    else:
        dataset = dataset.loc[dataset[presence_of[0]] == 1]
        for i in symptoms :
            if 1 in list(dataset[i]):
                frequency = round(dataset[i].value_counts()[1]/number_of_samples,5)
            else:
                frequency = 0
            symptom_distribution[i] = frequency
       
    return symptom_distribution
```

```python
general_nyumc_sd = list(get_general_distribution(data_nyumc,symptoms,[]).values())
print("NYUMC!")
print(general_nyumc_sd)
print("_______________________________")
general_gv_sd = list(get_general_distribution(data_gv,symptoms,[]).values())
print("GoViral")
print(general_gv_sd)
print("_______________________________")
general_fw_sd = list(get_general_distribution(data_fw,symptoms,[]).values())
print("FluWatch")
print(general_fw_sd)
print("_______________________________")
general_hk_sd = list(get_general_distribution(data_hk,symptoms,[]).values())
print("HongKong")
print(general_hk_sd)
print("_______________________________")
general_ht_sd = list(get_general_distribution(data_ht,symptoms,[]).values())
print("Hutterite")
print(general_ht_sd)
print("_______________________________")

general_sd = list(map(sum,zip(general_nyumc_sd,general_gv_sd,general_fw_sd,general_hk_sd,general_ht_sd)))

general_sd = [i/total_samples for i in general_sd]
print(general_sd)
```

### Get the weights for different levels of hierarchy

```python
#the weights for each level :
#for the first level depending on the source of the data
first_level = defaultdict()
first_level['clinically_collected'] = [round(size_nyumc/size_clinically_collected,2)]

first_level['individually_reported'] = [round(size_gv/size_individually_reported,2),
                                        round(size_fw/size_individually_reported,2)]

first_level['health_worker'] = [round(size_hk/size_health_worker,2),
                                round(size_ht/size_health_worker,2)]

print("Weights at the first level are : \n")
print(first_level)
```

```python
#at the second level all the weights are the same
second_level = defaultdict()
second_level['all'] = [round(size_clinically_collected/total_samples,2),
                       round(size_individually_reported/total_samples,2),
                       round(size_health_worker/total_samples,2)]

print("Weights at the second level are : \n")
print(second_level)
print("\n")
```
