### Determine the feature space

```python
common_features = []
all_features = []
```

```python
def get_common_features():
    common_features = list(set(symptoms['NYUMC']) & set(symptoms['GoViral']) & set(symptoms['FluWatch']) & set(symptoms['HongKong']) & set(symptoms['Hutterite']))
    return common_features
```

```python
common_features = get_common_features()
print("The common features in all the datasets are :\n")
for i in common_features:
    print(i)
print("Number of common features are :",len(common_features))
print("____________________________________")
```

There are four symptoms [fever, sorethroat, muscle, cough] that are common amongst all the datasets

```python
def get_all_features():
    all_features = list(set(symptoms['NYUMC']) | set(symptoms['GoViral']) | set(symptoms['FluWatch']) | set(symptoms['HongKong']) | set(symptoms['Hutterite']))
#     all_features.append('virus')
    return all_features
```

```python
all_features = get_all_features()
print("Features from all the datasets:\n")
for i in all_features:
    print(i)
print("Total number of symptoms across all the datasets :",len(all_features))
print("____________________________________")
```

```python
new_symptoms = defaultdict()
```

```python
def unify_feature_space(dataset,symptoms,target_variable,all_features):
    added_features = []
    for i in all_features:
        if i not in symptoms:
            dataset[i] = 0
            added_features.append(i)
    if target_variable == 'flu' :
        print("Target variable is Flu")
        dataset.rename(columns = {target_variable:'virus'},inplace = True)
    print("Number of columns now are : ",len(list(dataset.columns)))
    return dataset,added_features
```

```python
new_symptoms['NYUMC'] = symptoms['NYUMC']
data_nyumc,added_features_nyumc = unify_feature_space(data_nyumc,symptoms['NYUMC'],target_variable['NYUMC'],all_features)
print("Existing features in NYUMC :\n")
for i in symptoms['NYUMC']:
    print(i)
print("____________________________________")
print("Added features :\n")
for i in added_features_nyumc:
    print(i)
    new_symptoms['NYUMC'].append(i)
print("____________________________________")
data_nyumc.head()
```

### Ensure that only the requried features are present

```python
required_features = ['fever',
                     'sorethroat',
                     'cough',
                     'muscle',
                     'headache',
                     'fatigue',
                     'vomit',
                     'nausea',
                     'diarrhea',
                     'chills',
                     'sneeze',
                     'shortness of breath',
                     'phlegm',
                     'blockednose',
                     'earache',
                     'leg pain',
                     'runnynose',
                     'age 0-4',
                     'age 5-15',
                     'age 16-44',
                     'age 45-64',
                     'age 65+',
                     'male',
                     'female',
                     'virus']
```

```python
def feature_selection(dataset,required_features = required_features):
    dataset_features = list(dataset.columns)
    #check if we have missed out any symptoms
    not_present = [i for i in required_features if i not in dataset_features]
    print("The features we have missed are : ",not_present)
    excess = [i for i in dataset_features if i not in required_features]
    print("The excess features are : ",excess)
    dataset.drop(excess,axis = 1,inplace = True)
    dataset = dataset[required_features]
    return dataset
```

```python
data_nyumc = feature_selection(data_nyumc)
```

```python
data_gv = feature_selection(data_gv)
```

```python
data_fw = feature_selection(data_fw)
```

```python
data_hk = feature_selection(data_hk)
```

```python
data_ht = feature_selection(data_ht)
```

### Save the datasets with same feature space

```python
def save_file(dataframe_,filename_):
    print("Columns in this dataset are : ",list(dataframe_.columns))
    print(len(list(dataframe_.columns)))
    directory = "../Data/With_Demographics/"
    filename = directory+filename_
    if os.path.isfile(filename):
        print("File exists!")
    else:
        dataframe_.to_csv(filename,index=False)
```
